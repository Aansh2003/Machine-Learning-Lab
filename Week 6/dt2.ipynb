```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
col_names = ['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex', 'year' ]
pima = pd.read_csv("D:/210968186/penguins.csv", names=col_names)
pima = pima[1:].dropna()
pima = pima.replace(to_replace={'male':1,'female':0})
pima
```

```python
X = pima.drop(['body_mass_g','species','island'],axis=1)
Y = pima['body_mass_g']
#y = pima.label 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeRegressor()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("R2:",metrics.r2_score(y_test, y_pred))
```
