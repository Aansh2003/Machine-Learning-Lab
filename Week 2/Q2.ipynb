```python
# Question 2
column = False
for columns in df:
    if columns == 'calories':
        column = True
    if column:
        print('Five number summary for '+columns)
        print(df[columns].describe()[3:])
        print('\n')
```

```python
# Question 3
df= df.replace (-1,df.mean())
df
```

```python
# Question 4
boo = False
for value in df:
    if value == 'calories':
        boo = True
    if boo:
        mean = df[value].mean()
        std = df[value].std()
        median = df[value].median()
        
        upper = mean + 2.0*std
        lower = mean - 2.0*std
        
        print(upper)
        print(lower)
        
        df[value][df[value]>upper] = median
        df[value][df[value]<lower] = median
```

```python
# Question 5
column = False
for columns in df:
    if columns == 'calories':
        column = True
    if columns == 'rating':
        column = False
    if column:
        print('Five number summary for '+columns)
        print(df[columns].describe()[3:])
        print('\n')
```

```python
# Question 7
print("Correlation between calories and rating:")
print(df['calories'].corr(df['rating']))
print("They are correlated")
```

```python
# Question 8
labels, categories = pd.factorize(df["mfr"])
df["labels"] = labels
print('correlation:'+str(df["rating"].corr(df["labels"])))
print("They are not very related")
```
