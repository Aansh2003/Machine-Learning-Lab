```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('tested.csv')
df
```

```python
df = df = df.drop(['PassengerId','Name','Cabin','Ticket'],axis=1)
df
```

```python
df = df.dropna()
df = df.replace(to_replace={'male':1,'female':-1,'Q':3,'S':2,'C':1})
df
```

```python
df = df/df.max()
X = df.drop(['Survived'],axis=1)
Y = df['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

splits = [0.4,0.3,0.2,0.1]

for split in splits:
    X_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=split)
    X_train1 = X_train[:len(X_train)//3]
    X_train2 = X_train[len(X_train)//3:2*len(X_train)//3]
    X_train3 = X_train[2*len(X_train)//3:]
    y_train1 = y_train[:len(y_train)//3]
    y_train2 = y_train[len(y_train)//3:2*len(y_train)//3]
    y_train3 = y_train[2*len(y_train)//3:]
    
    m1 = DecisionTreeClassifier(max_depth=5)
    m2 = SVC(kernel='rbf')
    m3 = RandomForestClassifier()
    
    m1.fit(X_train1,y_train1)
    m2.fit(X_train2,y_train2)
    m3.fit(X_train3,y_train3)
    
    def predict(X_test):
        a = m1.predict(X_test)
        b = m2.predict(X_test)
        c = m3.predict(X_test)
        pred = []
        for i in range(len(a)):
            my_list = [a[i],b[i],c[i]]
            pred.append(max(set(my_list),key=my_list.count))
        return pred
    print('Split: ',split)
    print('Accuracy: ',accuracy_score(predict(x_test),y_test))
    print('\n')
    
    
```
