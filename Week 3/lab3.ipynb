```python
sns.histplot(df)
```

```python
sns.barplot(data=df)
```

```python
x = np.arange(0, 5*math.pi, 0.01)
y = np.sin(x)
plt.plot(x, y, color='green')
plt.show()
```

```python
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T
z = (np.sin(x **2) + np.cos(y **2) )

plt.axes(projection ="3d").plot_surface(x,y,z,color='green')
```
