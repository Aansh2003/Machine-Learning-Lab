```python
from sklearn.svm import SVR
import pandas as pd

housing = pd.read_csv("Housing.csv")
housing
```

```python
housing = housing.replace(to_replace={'yes':1,'no':0,'furnished':2,'unfurnished':0,'semi-furnished':1})
housing
```

```python
housing = housing/housing.max()
X = housing.drop(['price'],axis=1)
Y = housing['price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

splits = [0.6,0.7,0.8,0.9]

for split in splits:
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=1-split)
    regressor = SVR(kernel = 'rbf')
    regressor.fit(X_train,Y_train)

    y_pred = regressor.predict(X_test)
    print("Split size: ",split)
    print("R2 Score: ",r2_score(Y_test,y_pred))

```
