```python
import pandas as pd
import numpy as np
```

```python
x = []
y = []
for i in range(0,10):
    x.append(np.random.randint(0,100))
    y.append(np.random.randint(0,100))
```

```python
x,y
```

```python
y = np.asarray(y)
y = y.reshape(-1,1)
x = np.asarray(x)
x = x.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(x,y ,random_state=104,test_size=0.2, shuffle=True)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train,y_train)
y_new = regr.predict(X_train)
print(regr.coef_,regr.intercept_)
```

```python
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt

plt.plot(X_train,y_new)
```

```python
y_test_new = regr.predict(X_test)
plt.plot(X_test,y_test,color='green')
plt.plot(X_test,y_test_new)
plt.show()
```

```python
print("r2 score:"+str(metrics.r2_score(y_new,y_train)))
print("mean squared error:"+str(metrics.mean_squared_error(y_new,y_train)))
print("root mean squared error:"+str(metrics.mean_squared_error(y_new,y_train,squared=False)))
```
